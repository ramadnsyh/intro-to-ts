# Time Series Classical Components

A simple method of describing a time series is a classical decomposition: a proposition that a series can be decomposed into three main elements:

- Trend (Tt): long term movements in the mean
- Seasonal (St): repeated seasonal fluctuations (think: calendar or business cycles)
- Residuals (Et): Irregular components or random fluctuations not captured by the Trend and Seasonal

The idea of classical decomposition is to create separate models for each of the three elements in the aim of describing our series either additively:
Xt=Tt+St+Et

or multiplicatively: Xt=Tt⋅St⋅Et

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposed_df = df.copy()
result = seasonal_decompose(decomposed_df.births, model='additive', freq=12)
decomposed_df["trend"] = result.trend
decomposed_df["seasonal"] = result.seasonal
decomposed_df["resid"] = result.resid
decomposed_df["observed"] = result.observed

for col in ["observed", "trend", "seasonal", "resid"]:
    decomposed_df.plot(y=col, figsize=(12, 5), title=col)
```

# White Noise

# Random Walk

A random walk is different from a list of random numbers because the next value in the sequence is a modification of the previous value in the sequence.

Prediksi terbaik untuk random walk adalah harga kemarin

```python
from statsmodels.tsa.stattools import adfuller
```

```python
rwalk = pd.DataFrame(100 + np.cumsum(series))
rwalk.plot()
```

```python
plot_acf(series, lags=50);
```

## Dicky-Fuller Test

Sebuah uji yang bisa menunjukkan apakah suatu time-series itu random walk atau bukan

# Forecasting Method

- Naive
- Simple Moving Average
- Holt Winterz

## Naive Method

```python
df_forecasted = df.copy()
df_forecasted["naive_method"] = df.shift(1)
df_forecasted.head()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(df_forecasted.index, df_forecasted.births, label="Actual")
plt.plot(df_forecasted.index, df_forecasted.naive_method, label="Prediction")
plt.legend()
```

```python
df_forecasted.dropna(inplace=True)
mean_absolute_error(df_forecasted.births, df_forecasted.naive_method)
```

## Simple Moving Average

```python
df_forecasted = df.copy()
df_forecasted["t-1"] = df.shift(1)
df_forecasted["t-2"] = df.shift(2)
df_forecasted["t-3"] = df.shift(3)
df_forecasted["sma"] = df_forecasted.apply(lambda x: np.mean([x["t-1"], x["t-2"], x["t-3"]]), axis=1)
df_forecasted.head()
```

```python
plt.figure(figsize=(16, 5))
plt.plot(df_forecasted.index, df_forecasted.births, label="Actual")
plt.plot(df_forecasted.index, df_forecasted.sma, label="Prediction")
plt.legend()
```

```python
df_forecasted.dropna(inplace=True)
mean_absolute_error(df_forecasted.births, df_forecasted.sma)
```

## Simple Exponential Smoothing

This method is suitable for forecasting data with no clear trend or seasonal pattern

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
```

```python
model = SimpleExpSmoothing(df.births)
model_fit = model.fit(smoothing_level=0.5)
last_month_predicted = model_fit.predict(start=df.tail(100).index[0], end=df.tail(100).index[-1])
last_month_actual = df.tail(100)
mean_absolute_error(last_month_actual, last_month_predicted)
```

```python
plt.figure(figsize=(16, 5))
plt.plot(last_month_actual.index, last_month_actual, label="Actual")
plt.plot(last_month_predicted.index, last_month_predicted, label="Prediction")
plt.legend()
```

# Double Exponential Smoothing

Holt (1957) extended simple exponential smoothing to allow the forecasting of data with a trend

```python
model = ExponentialSmoothing(df.births, trend='multiplicative', seasonal=None)
model_fit = model.fit(smoothing_slope=0.3)
last_month_predicted = model_fit.predict(start=df.tail(100).index[0], end=df.tail(100).index[-1])
last_month_actual = df.tail(100)
mean_absolute_error(last_month_actual, last_month_predicted)
```

```python
plt.figure(figsize=(16, 5))
plt.plot(last_month_actual.index, last_month_actual, label="Actual")
plt.plot(last_month_predicted.index, last_month_predicted, label="Prediction")
plt.legend();
```

# Triplet Exponential Smoothing

Holt (1957) and Winters (1960) extended Holt’s method to capture seasonality

```python
model = ExponentialSmoothing(df.births, trend='mul', seasonal="add", seasonal_periods=12)
model_fit = model.fit(smoothing_seasonal=0.65)
last_month_predicted = model_fit.predict(start=df.tail(100).index[0], end=df.tail(100).index[-1])
last_month_actual = df.tail(100)
mean_absolute_error(last_month_actual, last_month_predicted)
```

```python
plt.figure(figsize=(16, 5))
plt.plot(last_month_actual.index, last_month_actual, label="Actual")
plt.plot(last_month_predicted.index, last_month_predicted, label="Prediction")
plt.legend()
```

##  Stationarity and differencing

A stationary time series is one whose properties do not depend on the time at which the series is observed.14 Thus, time series with trends, or with seasonality, are not stationary — the trend and seasonality will affect the value of the time series at different times. On the other hand, a white noise series is stationary — it does not matter when you observe it, it should look much the same at any point in time.

```python
from statsmodels.tsa.stattools import adfuller
```

## AR: Autoregressive models

In a multiple regression model, we forecast the variable of interest using a linear combination of predictors. In an autoregression model, we forecast the variable of interest using a linear combination of past values of the variable

```python
from statsmodels.tsa.arima_model import ARMA
```

```python
AR = ARMA(births, order=(1, 0))
model = AR.fit()
print(f"AIC: {model.aic}, BIC: {model.bic}")
```

```python
model.plot_predict(end=len(births) + 24);
```

```python
AR = ARMA(births, order=(2, 0))
model = AR.fit()
print(f"AIC: {model.aic}, BIC: {model.bic}")
```

```python
model.plot_predict(end=len(births) + 24);
```

## ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
arima = ARIMA(df.births, order=(1, 1, 4))
model = arima.fit()
last_month_predicted = model.predict(start=df.tail(100).index[0], end=df.tail(100).index[-1], typ="levels")
last_month_actual = df.tail(100)
print(f"AIC: {model.aic}, BIC: {model.bic}")
print(f"MAE: {mean_absolute_error(last_month_actual, last_month_predicted)}")
```

```python
model.plot_predict(end=len(births) + 24);
```

```python
plt.figure(figsize=(16, 5))
plt.plot(last_month_actual.index, last_month_actual, label="Actual")
plt.plot(last_month_predicted.index, last_month_predicted, label="Prediction")
plt.legend();
```
